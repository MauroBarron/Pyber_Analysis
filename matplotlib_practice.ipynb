```python
# Import dependencies (the matplotlib library plus pyplot module.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create first python pyplot plot. yay. 
plt.plot(x_axis, y_axis)
```

```python
## Now we create line chart with ax.plot()
 # This is an object oriented method of plotting. 
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# Now we practice for understanding syntax 
# figure, axis = plt.subplots() # figure is first variable, axis is second
# axis.plot(x_axis, y_axis)    
```

```python
# practice some more. break out components of the command
# create plot with the ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
    
```

```python
## Annotating graph using MATHLAB
#  create the plot and a label
plt.plot(x_axis, y_axis, label = "Boston")  # label is for the legend
         
# Label the axis
plt.xlabel("Date")
plt.ylabel ("Fare($)")
         
# Set the Y limit between 0 and 45
plt.ylim(0,45)
         
# plot title
plt.title("PyBer Fare by Month")
         
# now a legend
plt.legend()
```

```python
## Adding more visual details to the chart.
#  Create the plot
plt.plot(x_axis, y_axis,  marker="*", color="green", linewidth=2, label = "Boston")  # label is for the legend
# Label the axis
plt.xlabel("Date")
plt.ylabel ("Fare($)")
# Set the x limit between January and June
#plt.xlim("Jan", "June")
plt.xlim(0, 5)
# Set the Y limit between 0 and 45
plt.ylim(0,45)
# plot title
plt.title("PyBer Fare by Month")
# add a grid
plt.grid()
# now a legend
plt.legend()
```

```python
## now practice figure sizing.
# Control figure size
#plt.figure(figsize=(20,10))
plt.subplots(figsize=(20,10))
#  Annotate the plot
plt.plot(x_axis, y_axis,  marker="*", color="green", linewidth=2, label = "Boston")  # label is for the legend
# Label the axis
plt.xlabel("Date")
plt.ylabel ("Fare($)")
# Set the x limit between January and June
#plt.xlim("Jan", "June")
plt.xlim(0, 5)
# Set the Y limit between 0 and 45
plt.ylim(0,45)
# plot title
plt.title("PyBer Fare by Month")
# add a grid
plt.grid()
# now a legend
plt.legend()
```

```python
## Skill Drill. Annotate the chart using Object Oriented Interface.

# control the size
fig, ax = plt.subplots(figsize=(20, 10))

ax.set_title("PyBer Fare by Month")

# configure the properties and label
ax.plot(x_axis, y_axis, color="green",linewidth=2, label = "Boston", marker ="D")
ax.legend()

# Label the axis
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Set the x limit between January and June
#ax.set_xlim(0,5)
# Set the Y limit between 0 and 45
ax.set_ylim(0,45)
 
# add a grid and legend
ax.legend()
ax.grid()

#  save the figure to drive.
#**plt.savefig("C:\Users\Marishka\Documents\DABootCamp\W6-PyBerWithMatplotlib\Pyber_Analysis\514SkillDrill.png")
plt.savefig("514SkillDrill.png")

```

```python
### Now we learn to make bar charts using MATLAB!
#   Create the plot
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Label the axis
plt.xlabel("Date")
plt.ylabel ("Fare($)")

# title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
## Pracice Horizontal bar charts.

# create the plot
plt.barh(x_axis, y_axis, color='magenta', label = "Boston")  # plots an HBAR

# plot title
plt.title("PyBer Fare by Month")

# Label the axis
plt.xlabel("Fare($)")
plt.ylabel ("Date")

# Use the gca() method to get current axes.Chain with the invert_yaxis method 
plt.gca().invert_yaxis() 

# # now a legend
plt.legend()
```

```python
# Skill Drill 5.1.7

plt.scatter(y_axis, x_axis, color="red", label = 'Chicago', s=y_axis, marker = 'd')
plt.legend()

# plot title
plt.title("PyBer Fare by Month")

# Label the axis
plt.xlabel("Fare($)")
plt.ylabel ("Date")

# Set the x limit between 0 and 45
plt.xlim(0,45)

# Use the gca() method to get current axes.Chain with the invert_yaxis method 
plt.gca().invert_yaxis() 

#  save the figure to drive.
plt.savefig("517SkillDrill.png")
```

```python
## Now we can create a bubble Cha

## make the marker exponentially larger base on sixe of Y
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis,  label = 'Chicago', s=y_axis_larger)
plt.legend()

# plot title
plt.title("PyBer Fare by Month")

# Label the axis
plt.ylabel("Fare($)")
plt.xlabel ("Date")

# Set the x limit between 0 and 45
plt.ylim(0,45)

# Use the gca() method to get current axes.Chain with the invert_yaxis method 
#plt.gca().invert_yaxis() 

#  save the figure to drive.
plt.savefig("517BubbleChart.png")
```

```python
## now practice list comprehension. Add the loop with the scatterplot function

# list_name- (expresssion for item in list if conditional)
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
## Scatter plot using object-oriented interface.
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
## now a bubble chart using object-oriented
fig, ax=plt.subplots()
ax.scatter(y_axis, x_axis, c = "skyblue",  s = [i * 5 for i in y_axis], edgecolors ="black", linewidth=2, label = "Boston", alpha=0.8)

## note - changing the transperancy to below zero results in nothing displayed on my pc.
#ax.scatter(x_axis, y_axis, alpha=0.1)

# #Set a title
ax.set_title("PyBer Fare by Month")

# # Label the axis
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# #Invert the Y-axis so January is on top
plt.gca().invert_yaxis() 

# Set the X limit between 0 and 45
ax.set_xlim(0,45)

# # Add legend
ax.legend()

# #  save the figure to drive.
plt.savefig("517SkillDrillBubbleObjectOriented.png")
```

```python
##  Pie charts
explode_values = (0,0,0,0,0,0,0.25,0,0,0,0,0) # designates 7th value to explore
plt.subplots(figsize=(8,8))  # size of the pie chart
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct="%.1f%%") # more customization
plt.savefig("518_PieChart.png")  # save to disc
plt.show # show the chart.
```

```python
# Assign 12 colors, one for each month. # taken from course page.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
## Now we create a pie chare with object-oriented code
fig, ax= plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
## Skill Drill 5.1.8. Pie charts with object oriented code.
fig, ax= plt.subplots(figsize=(8,8))  # creates pie with size 8x8
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, # y_axis as values
    labels=x_axis
    ,autopct='%.1f%%' # percentage to one decimal point
    ,explode = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)    
    ,shadow = True
    ,startangle = 95
    ,counterclock = False
    ,colors = colors
    )
ax.set_title("PyBer Fare by Month") #Set a title
#ax.legend() # show legend
plt.savefig("518SkillDrillPieChartsObjectOriented.png") # save chart to drive
plt.show()
```
