```python
###  Now into the module. Start by setting magic command.
##  Create a line chart using MATLAB
## The magic command. % sign generates back end processeor action to prep 
 # environment for desired usage.
 # this specific magic command produces plots inline within the Jupyter Notebook. 
%matplotlib inline
```

```python
# Import dependencies (the matplotlib library plus pyplot module.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create first python pyplot plot. yay. 
plt.plot(x_axis, y_axis)
```

```python
## Now we create line chart with ax.plot()
 # This is an object oriented method of plotting. 
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# Now we practice for understanding syntax 
# figure, axis = plt.subplots() # figure is first variable, axis is second
# axis.plot(x_axis, y_axis)    
```

```python
# practice some more. break out components of the command
# create plot with the ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
    
```

```python
# more practice wioth ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
## Annotating graph using MATHLAB
#  create the plot and a label
plt.plot(x_axis, y_axis, label = "Boston")  # label is for the legend
         
# Label the axis
plt.xlabel("Date")
plt.ylabel ("Fare($)")
         
# Set the Y limit between 0 and 45
plt.ylim(0,45)
         
# plot title
plt.title("PyBer Fare by Month")
         
# now a legend
plt.legend()
```

```python
## Adding more visual details to the chart.
#  Create the plot
plt.plot(x_axis, y_axis,  marker="*", color="green", linewidth=2, label = "Boston")  # label is for the legend
# Label the axis
plt.xlabel("Date")
plt.ylabel ("Fare($)")
# Set the x limit between January and June
#plt.xlim("Jan", "June")
plt.xlim(0, 5)
# Set the Y limit between 0 and 45
plt.ylim(0,45)
# plot title
plt.title("PyBer Fare by Month")
# add a grid
plt.grid()
# now a legend
plt.legend()
```

```python
## now practice figure sizing.
# Control figure size
#plt.figure(figsize=(20,10))
plt.subplots(figsize=(20,10))
#  Annotate the plot
plt.plot(x_axis, y_axis,  marker="*", color="green", linewidth=2, label = "Boston")  # label is for the legend
# Label the axis
plt.xlabel("Date")
plt.ylabel ("Fare($)")
# Set the x limit between January and June
#plt.xlim("Jan", "June")
plt.xlim(0, 5)
# Set the Y limit between 0 and 45
plt.ylim(0,45)
# plot title
plt.title("PyBer Fare by Month")
# add a grid
plt.grid()
# now a legend
plt.legend()
```

```python
## Skill Drill. Annotate the chart using Object Oriented Interface.

# control the size
fig, ax = plt.subplots(figsize=(20, 10))

ax.set_title("PyBer Fare by Month")

# configure the properties and label
ax.plot(x_axis, y_axis, color="green",linewidth=2, label = "Boston", marker ="D")
ax.legend()

# Label the axis
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Set the x limit between January and June
#ax.set_xlim(0,5)
# Set the Y limit between 0 and 45
ax.set_ylim(0,45)
 
# add a grid and legend
ax.legend()
ax.grid()

#  save the figure to drive.
#**plt.savefig("C:\Users\Marishka\Documents\DABootCamp\W6-PyBerWithMatplotlib\Pyber_Analysis\514SkillDrill.png")
plt.savefig("514SkillDrill.png")

```

```python
### Now we learn to make bar charts using MATLAB!
#   Create the plot
plt.bar(x_axis, y_axis, color="green", label="Boston")
# Label the axis
plt.xlabel("Date")
plt.ylabel ("Fare($)")

# title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
## Pracice Horizontal bar charts.

# create the plot
plt.barh(x_axis, y_axis, color='magenta', label = "Boston")  # plots an HBAR

# plot title
plt.title("PyBer Fare by Month")

# Label the axis
plt.xlabel("Fare($)")
plt.ylabel ("Date")

# Use the gca() method to get current axes.Chain with the invert_yaxis method 
plt.gca().invert_yaxis() 

# # now a legend
plt.legend()
```

```python
## Create a veritcal bar chart using object-oriented method
#  Uses the ax.bar method
# Create th plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="black", label="New York")
```

```python
#  Create a horizontal bar. Still using object oriented method
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```
